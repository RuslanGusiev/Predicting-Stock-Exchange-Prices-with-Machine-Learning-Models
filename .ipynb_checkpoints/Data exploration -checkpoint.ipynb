```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_excel.html
data_house_1 = pd.read_excel("data/Private Home_1_2011_June_03_to_17.xls", sheet_name='Trend', header=6)
```

```python
data_house_1[['Frequency','W1','Wh1']] = data_house_1[['Frequency','W1','Wh1']].apply(pd.to_numeric,errors='coerce')
house_1 = data_house_1[['Frequency','W1','Wh1', 'Date', 'Time']] 
house_1
```

```python
house_1["isnull"] = house_1.isnull().any(axis=1)
missing_values = house_1[house_1['isnull'] == True]
missing_values # the data set in clean there are no missing values 
```

```python
house_1 = house_1.drop(['isnull'], axis=1)
house_1.isna().sum()
```

```python
house_1['DateTime'] = pd.to_datetime(house_1.Date + ' ' + house_1.Time)
house_1['Time_only'] = pd.to_datetime(house_1.Time)
```

```python
house_1['hour'] = house_1.DateTime.dt.hour
house_1['day'] = house_1.DateTime.dt.day_name()
house_1['min'] = house_1.DateTime.dt.minute
#house_1['5_min_perday']= house_1.hour + house_1.min
house_1.groupby(['Time_only']).mean()
```

```python
#sns.set_style("darkgrid", {"axes.facecolor": ".9"})
sns.set(rc={'figure.figsize':(16,5)})
sns.set_style('ticks')
```
