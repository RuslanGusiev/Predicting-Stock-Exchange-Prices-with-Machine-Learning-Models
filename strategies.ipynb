```python
date = [t for t in AAPL.date] 
close = [t for t in AAPL.close]
```

```python
buy = 0 # price
ind1 = 0 # indicator for buying
ind2 = 0 # indicator for selling
result = 0

for i in range(len(date)):
    if (close[i-3] > close[i-2]) and (close[i-3] > close[i-1]) and (close[i-3] > close[i]) and (close[i-3] > close[i-4]) and (close[i-3] > close[i-5]) and (close[i-3] > close[i-6]):
        ind1 = close[i-3]
    if (close[i-3] < close[i-2]) and (close[i-3] < close[i-1]) and (close[i-3] < close[i]) and (close[i-3] < close[i-4]) and (close[i-3] < close[i-5]) and (close[i-3] < close[i-6]):
        ind2 = close[i-3]

    if (buy ==0) and (close[i] > ind1):
        buy = close[i]
    if (buy != 0) and (close[i]< ind2):
        result  += close[i] - buy
        print(result)
        print(round((close[i] - buy)/buy*100, 2), '%')
        buy = 0
        
        
    
```
